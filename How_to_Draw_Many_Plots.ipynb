# 1 MATLAB-like API

# 2 fig.add_axes

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,3)
y1 = x ** 2
y2 = np.sqrt(x)

if 0:
    fig = plt.figure()
    ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
    ax.plot(x, y1)
elif 1:
    fig = plt.figure()
    ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
    ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes
    ax1.plot(x, y1) # main figure
    ax2.plot(y1, x) # insert

plt.show()
```

# 3 fig.add_subplot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,1)
y1 = x ** 2
y2 = np.sqrt(x)

fig = plt.figure(figsize=(16, 12))

# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots_adjust.html
fig.subplots_adjust(hspace=0.1, wspace=0.3)

# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.suptitle.html
fig.suptitle("Iterative_Policy_Evaluation for V", fontsize=16)

for s in range(11):
    ax = fig.add_subplot(3, 4, s + 1)
    if s%2 == 0:
        ax.plot(y1, label="state {}".format(str(s)))
    else:
        ax.plot(y2, label="state {}".format(str(s)))
    ax.legend(loc='upper right')
    ax.set_ylim([-1.1, 1.1])

plt.show()
```

# 5 plt.subplot2grid

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,3)
y1 = x ** 2
y2 = np.sqrt(x)

fig = plt.figure()
ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2,0))
ax5 = plt.subplot2grid((3,3), (2,1))
fig.tight_layout()

ax1.plot(x,y1)
ax2.plot(y1,x)
ax3.plot(x,y1)
ax4.plot(x,y2)
ax5.plot(y2,x)

plt.show()
```

# 6 gridspec.GridSpec

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

x = np.linspace(0,3)
y1 = x ** 2
y2 = np.sqrt(x)

fig = plt.figure()
gs = gridspec.GridSpec(2, 3, height_ratios=[2,1], width_ratios=[1,2,1])
for i, g in enumerate(gs):
    ax = fig.add_subplot(g)
    if i == 0:
        ax.plot(x,y1)
    elif i == 1: 
        ax.plot(y1,x)
    elif i == 2:
        ax.plot(x,y1)
    elif i == 3:
        ax.plot(x,y2)
    elif i == 4:
        ax.plot(y2,x)
    elif i == 5:
        ax.plot(x,y2)
fig.tight_layout()

plt.show()
```
