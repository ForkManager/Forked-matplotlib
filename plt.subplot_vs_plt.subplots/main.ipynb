```python
%matplotlib inline
import matplotlib.pyplot as plt

days = [10,11,12,13,14,15,16]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]

plt.subplot(2,2,1)
plt.plot(days, max_t, linestyle='--', color='k', marker='o', label='max')
plt.legend(loc='lower left', fontsize='large', shadow=True)
plt.grid(True)
plt.title('Seoul Max Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature')

plt.subplot(2,2,2)
plt.plot(days, avg_t, color='#2714e7', label='average')
plt.legend(loc='lower left', fontsize='large', shadow=True)
plt.grid(True)
plt.title('Seoul Average Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature')

plt.subplot(2,2,3)
plt.plot(days, min_t, linestyle='--', color='k', marker='o', label='min')
plt.legend(loc='lower left', fontsize='large', shadow=True)
plt.grid(True)
plt.title('Seoul Min Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature')

plt.tight_layout()

plt.savefig('plot.png')

plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

days = [10,11,12,13,14,15,16]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]

fig, axes = plt.subplots(nrows=3, ncols=1, sharex=True)

axes[0].plot(days, max_t, linestyle='--', color='k', marker='o', label='max')
axes[0].legend(loc='lower left', fontsize='large', shadow=True)
axes[0].grid(True)
axes[0].set_title('Seoul')
#axes[0].set_xlabel('Date')
axes[0].set_ylabel('Temperature')

axes[1].plot(days, avg_t, color='#2714e7', label='average')
axes[1].legend(loc='lower left', fontsize='large', shadow=True)
axes[1].grid(True)
#axes[1].set_title('Seoul Average Temperature')
#axes[1].set_xlabel('Date')
axes[1].set_ylabel('Temperature')

axes[2].plot(days, min_t, linestyle='--', color='k', marker='o', label='min')
axes[2].legend(loc='lower left', fontsize='large', shadow=True)
axes[2].grid(True)
#axes[2].set_title('Seoul Min Temperature')
axes[2].set_xlabel('Date')
axes[2].set_ylabel('Temperature')

plt.tight_layout()

plt.show()

fig.savefig('plot.png')
```
