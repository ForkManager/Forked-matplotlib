# figsize, dpi 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 10)
y = x ** 2

if 0:
    fig = plt.figure()
elif 1:
    # figsize: a tuple of the width and height of the figure in inches
    # dpi: dots-per-inch (pixel per inch).
    fig = plt.figure(figsize=(8,4), dpi=100)
    
plt.plot(x, y)

# What formats are available and which ones should be used for best quality?
# Matplotlib can generate high-quality output in a number formats, 
# including PNG, JPG, EPS, SVG, PGF and PDF. 
# For scientific papers, I recommend using PDF whenever possible. 
# (LaTeX documents compiled with pdflatex can include PDFs 
# using the includegraphics command). 
# In some cases, PGF can also be good alternative.
if 0:
    fig.savefig("img/filename.png")
elif 1:
    fig.savefig("img/filename.png", dpi=100)

plt.show()
```

# label, legend, loc, shadow, fontsize

```python
import matplotlib.pyplot as plt

days = [0,1,2,3,4,5,6]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]

plt.figure()

if 0:
    plt.plot(days, max_t)
    plt.plot(days, min_t)
    plt.plot(days, avg_t)
elif 0:
    plt.plot(days, max_t, label="max")
    plt.plot(days, min_t, label="min")
    plt.plot(days, avg_t, label="average")
    plt.legend()
elif 0:
    plt.plot(days, max_t)
    plt.plot(days, min_t)
    plt.plot(days, avg_t)
    plt.legend(["max", "min", "average"])
elif 1:
    # https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html
    plt.plot(days, max_t, '--r', label="max")
    plt.plot(days, min_t, '--r', label="min")
    plt.plot(days, avg_t, '-b', label="average")
    plt.legend(loc='upper right', shadow=True, fontsize='large')
    plt.grid(True)
    
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.show()
```

# Formatting text: LaTeX, fontsize, font family

The figure above is functional, but it does not (yet) satisfy the criteria for a figure used in a publication. First and foremost, we need to have LaTeX formatted text, and second, we need to be able to adjust the font size to appear right in a publication.

Matplotlib has great support for LaTeX. All we need to do is to use dollar signs encapsulate LaTeX in any text (legend, title, label, etc.). For example, `"$y=x^3$"`.

But here we can run into a slightly subtle problem with LaTeX code and Python text strings. In LaTeX, we frequently use the backslash in commands, for example `\alpha` to produce the symbol $\alpha$. But the backslash already has a meaning in Python strings (the escape code character). To avoid Python messing up our latex code, we need to use "raw" text strings. Raw text strings are prepended with an '`r`', like `r"\alpha"` or `r'\alpha'` instead of `"\alpha"` or `'\alpha'`:

jrjohansson [github](https://github.com/jrjohansson/scientific-python-lectures/blob/master/Lecture-4-Matplotlib.ipynb)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 10)
y1 = x**2
y2 = x**3

fig, ax = plt.subplots()

ax.plot(x, y1, label=r"$y = \alpha^2$")
ax.plot(x, y2, label=r"$y = \alpha^3$")
ax.legend() 
ax.set_xlabel(r'$\alpha$', fontsize=18)
ax.set_ylabel(r'$y$', fontsize=18)
ax.set_title('title')

plt.show()
```

# color, alpha, linewidth, linetype, marker, markersize, markerfacecolor, markeredgewidth, markeredgecolor

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 10)

fig, ax = plt.subplots(dpi=400)

if 0:
    # MATLAB style line color and style 
    ax.plot(x, x**2, 'b.-') # blue line with dots
    ax.plot(x, x**3, 'g--') # green dashed line
elif 0:
    # https://htmlcolorcodes.com/color-picker/
    ax.plot(x, x+1, color="red", alpha=0.5) # half-transparant red
    ax.plot(x, x+2, color='#D63FB1')
    #ax.plot(x, x+2, color="#1155dd")       # RGB hex code for a bluish color
    ax.plot(x, x+3, color="#15cc55")        # RGB hex code for a greenish color
elif 0:
    ax.plot(x, x+1, color="blue", linewidth=0.25)
    ax.plot(x, x+2, color="blue", linewidth=0.50)
    ax.plot(x, x+3, color="blue", linewidth=1.00)
    ax.plot(x, x+4, color="blue", linewidth=2.00)
elif 0:
    # possible linestype options -, --, -., :, steps
    ax.plot(x, x+1, c="r", lw=2, linestyle='-')
    ax.plot(x, x+2, c="r", lw=2, ls='--')
    ax.plot(x, x+3, c="r", lw=2, ls='-.')
    ax.plot(x, x+4, c="r", lw=2, ls=':')
elif 0:
    # custom dash
    line, = ax.plot(x, x+1, color="black", lw=1.50)
    line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...
elif 0:
    # https://matplotlib.org/api/markers_api.html
    # possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
    ax.plot(x, x+1, c="g", lw=2, ls='--', marker='+')
    ax.plot(x, x+2, c="g", lw=2, ls='--', marker='o')
    ax.plot(x, x+3, c="g", lw=2, ls='--', marker='s')
    ax.plot(x, x+4, c="g", lw=2, ls='--', marker='p')
elif 1:
    # marker size and color
    ax.plot(x, x+1, c="b", lw=1, ls='-', marker='o', markersize=2)
    ax.plot(x, x+2, c="b", lw=1, ls='-', marker='o', markersize=4)
    ax.plot(x, x+3, c="b", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
    ax.plot(x, x+4, c="b", lw=1, ls='-', marker='s', markersize=8, 
            markerfacecolor="yellow", markeredgewidth=2, markeredgecolor="blue")

plt.show()
```

# Plot range

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 10)
y = x ** 2

if 0:
    plt.figure()
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.xlim([2, 5])
    plt.ylim([0, 60])
    plt.title('title')    
elif 1:
    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_xlim([2, 5])
    ax.set_ylim([0, 60])
    ax.set_title('title')
    
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 10)

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_xlim([2, 5])
axes[2].set_ylim([0, 60])
axes[2].set_title("custom axes range")

plt.show()
```

# Logarithmic scale

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 10)

fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)")

plt.show()
```
