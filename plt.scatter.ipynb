```python
# https://github.com/joeyajames/Python/blob/master/Matplotlib/Python%20Matplotlib%20Tutorial.ipynb

import numpy as np
import matplotlib.pyplot as plt

# number of data
N = 50

# 4-dim data
x = np.random.uniform(0., 1., (N,))
y = np.random.uniform(0., 1., (N,))
z = np.random.uniform(0., 1., (N,))
info = (30 * np.random.uniform(0., 1., (N,)))**2   # 0 to 15 point radii

plt.scatter(x, y, c=z, s=info, alpha=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time 

# number of data
N = 100000

# 4-dim data
x = np.random.uniform(0., 1., (N,))
y = np.random.uniform(0., 1., (N,))
z = np.random.uniform(0., 1., (N,))
info = (30 * np.random.uniform(0., 1., (N,)))**2   # 0 to 15 point radii

tic = time.time()
plt.scatter(x, y, c=z, s=info, alpha=0.5)
plt.show()
toc = time.time()
print(toc-tic)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time 

# number of data
N = 100000

# 4-dim data
x = np.random.uniform(0., 1., (N,))
y = np.random.uniform(0., 1., (N,))
z = np.random.uniform(0., 1., (N,))
info = (30 * np.random.uniform(0., 1., (N,)))**2   # 0 to 15 point radii

tic = time.time()
plt.plot(x, y, 'or', markersize=12, alpha=0.5)
plt.show()
toc = time.time()
print(toc-tic)
```
