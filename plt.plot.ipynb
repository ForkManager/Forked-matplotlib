```python
import matplotlib.pyplot as plt

days = [0,1,2,3,4,5,6]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]

plt.figure()

if 0:
    plt.plot(avg_t)
elif 0:
    plt.plot(days, avg_t)
elif 0:
    plt.plot(days, avg_t, '-oy')
elif 0:
    plt.plot(days, max_t)
    plt.plot(days, min_t)
    plt.plot(days, avg_t)
elif 0:
    plt.plot(days, max_t, label="max")
    plt.plot(days, min_t, label="min")
    plt.plot(days, avg_t, label="average")
    plt.legend()
elif 0:
    plt.plot(days, max_t, '--r', label="max")
    plt.plot(days, min_t, '--r', label="min")
    plt.plot(days, avg_t, '-b', label="average")
    plt.legend()
elif 0:
    # https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html
    plt.plot(days, max_t, '--r', label="max")
    plt.plot(days, min_t, '--r', label="min")
    plt.plot(days, avg_t, '-b', label="average")
    plt.legend(loc='upper right', shadow=True, fontsize='large')
elif 1:
    plt.plot(days, max_t, '--r', label="max")
    plt.plot(days, min_t, '--r', label="min")
    plt.plot(days, avg_t, '-b', label="average")
    plt.legend(loc='upper right', shadow=True, fontsize='large')
    plt.grid(True)
    
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.show()
```
