```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

plt.xkcd()

n = 7000

SPY = pd.read_csv('dow30/SPY.csv').set_index('Date')
SPY.index = pd.to_datetime(SPY.index)
DATE = SPY.index[:n]
SPY = (SPY['Adj Close'].loc[DATE] / SPY['Adj Close'].loc[DATE[0]]) * 100

AAPL = pd.read_csv('dow30/AAPL.csv').set_index('Date')
AAPL.index = pd.to_datetime(AAPL.index)
AAPL = (AAPL['Adj Close'].loc[DATE] / AAPL['Adj Close'].loc[DATE[0]]) * 100

WMT = pd.read_csv('dow30/WMT.csv').set_index('Date')
WMT.index = pd.to_datetime(WMT.index)
WMT = (WMT['Adj Close'].loc[DATE] / WMT['Adj Close'].loc[DATE[0]]) * 100

plt.plot(DATE, SPY, label='SPY')
plt.plot(DATE, AAPL, label='AAPL')
plt.plot(DATE, WMT, label='WMT')

plt.xlabel('DATE')
plt.ylabel('STOCK PRICE')
plt.title('STOCK PRICE (STARTING AT 100)')

plt.legend()
plt.show()
```
